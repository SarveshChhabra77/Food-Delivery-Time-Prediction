```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
```

```python
df=pd.read_csv('D:\Food Delivery Time Prediction\Data\RawData\Food_Delivery_Times.csv')
```

```python
## droping the null values 
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
categorical_features=[feature for feature in df.columns if df[feature].dtype=='O']
numerical_features=[feature for feature in df.columns if df[feature].dtype!='O']
```

```python
print(f'Numerical Columns : {numerical_features}')
print(f'Categorical Columns : {categorical_features}')
```

```python
plt.figure(figsize=(15, 15))  # Wider and shorter for 3 plots
plt.suptitle('Univariate Analysis of Numerical Features',
             fontsize=20, fontweight='bold', alpha=0.8, y=1.02)

for i in range(len(numerical_features)):
    plt.subplot(5, 3, i+1)  # 5 row, 3 columns
    sns.kdeplot(x=df[numerical_features[i]], shade=True, color='r')
    plt.xlabel(numerical_features[i])
    plt.tight_layout()
```

 All numerical features follows normal distribution 

```python
plt.figure(figsize=(15,15))
plt.suptitle('Univariate Analysis of Categorical Features',
             fontsize=20, fontweight='bold', alpha=0.8, y=1.02)

for i in range(len(categorical_features)):
    plt.subplot(4,3,i+1)
    sns.countplot(x=df[categorical_features[i]],color='b')
    plt.xlabel(categorical_features[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
    
```

From the above graph we have observed that from weather feature most of the people order food when the weather is clear, from Traffic_level we observed that mostly people ordered food where traffic level is low or medium, from Time_of_day we observed that mostly people ordered food during day time,and most of the order is delivered from Bike
